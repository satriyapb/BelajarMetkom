```python
# Mengimport module yang digunakan dan mendefinisikan dengan alias yang lebih singkat
import numpy as np                             # Digunakan untuk membuat objek kelas array tunggal dan multidimensi.
import pylab as plt                            # Digunakan untuk visualisasi data

# Properti atau komponen dari fungsi signal
A=1                                            # Amplitudo
omega=1                                        # Frekuensi anguler 
theta=np.pi/2                                  # Sudut fase
T = 2*np.pi/omega                              # Periode signal

# Interval waktu
dt=0.01                                        # Resolusi grafik
t=np.arange(0,2*np.pi+dt,dt)

# Definisi fungsi dari signal
x=A*np.cos(omega*t)+A*np.sin(omega*t+theta)

# Plotting data dari signal
plt.plot(t,x)                                  # Data yang akan di plot yaitu data "x" terhadap perubahan "t"
plt.xlabel("t")                                # Label data sumbu x
plt.ylabel("x(t)")                             # Label data sumbu y

# Judul grafik
plt.suptitle("Grafik 1. Sinyal Sinusoidal", fontsize = 20, y=1)

# Menampilkan grafik dari data yang diplot
plt.show()

```

```python
# Mengimport module yang digunakan dan mendefinisikan dengan alias yang lebih singkat
import numpy as np                             # Digunakan untuk membuat objek kelas array tunggal dan multidimensi.
import pylab as plt                            # Digunakan untuk visualisasi data

# Interval waktu
dt=0.01                                        # Resolusi grafik
t=np.arange(-5,5+dt,dt)

# Menentukan nilai variabel konstan dari persamaan eksponen
b=-1/0.4

# Mendefinisikan fungsi dari signal
x=1/np.sqrt(2*np.pi)*np.exp(b*t**2)

# Plotting data dari signal
plt.plot(t, x,)                            # Data yang akan di plot yaitu data "x" terhadap perubahan "t" 
plt.xlabel('t')                                # Label dari sumbu x
plt.ylabel('x(t)')                             # Label dari sumbu y

# Judul grafik
plt.suptitle("Grafik 2. Sinyal Eksponensial", fontsize = 20, y=1)

# Menampilkan grafik dari data yang diplot
plt.show()
```

```python
# Mengimport module yang digunakan dan mendefinisikan dengan alias yang lebih singkat
import numpy as np                             # Digunakan untuk membuat objek kelas array tunggal dan multidimensi.
import pylab as plt                            # Digunakan untuk visualisasi data

# Interval waktu
dt=0.01                                        # Resolusi grafik
t=np.arange(-5,10+dt,dt)

# Mendefinisikan signal dari fungsi heaviside
h=np.heaviside(t+2,1)

# Mendefinisikan fungsi dari signal x(t)
x=np.exp(-t)*h

# Melakukan plotting data sinyal
plt.plot(t, x,)                            # Data yang akan di plot yaitu data "x" terhadap perubahan "t"
plt.xlabel('t')                                # Label dari sumbu x
plt.ylabel('x(t)')                             # Label dari sumbu y
# Memberikan judul grafik
plt.suptitle("Grafik 3. Heaviside Step Function", fontsize = 20, y=1)

# Menampilkan grafik
plt.show() 
```

```python
# Mengimport module yang digunakan dan mendefinisikan dengan alias yang lebih singkat
import numpy as np                             # Digunakan untuk membuat objek kelas array tunggal dan multidimensi.
import pylab as plt                            # Digunakan untuk visualisasi data

# Interval waktu dt untuk resolusi grafik
dt=0.1      

# Interval waktu pertama 
t1=np.arange(-5,1.9+dt,dt)                     # Saat t = -5 hingga t= 1.9

# Waktu kedua
t2=[2.]                                        # Saat t = 2

# Interval waktu ketiga
t3=np.arange(2.1,10+dt,dt)                     # Saat t = 2.1 hingga t = 10

# Definisi nilai fungsi d(t) saat t1
d1=np.zeros_like(t1)                           # d(t) bernilai 0

# Definisi nilai fungsi d(t) saat t2
d2=[1.]                                        # d(t) bernilai 1

# Definisi nilai fungsi d(t) saat t3
d3=np.zeros_like(t3)                           # d(t) bernilai 0

# Gabungan dari variabel t dan d
t = np.concatenate((t1,t2,t3))
d = np.concatenate((d1,d2,d3))

# Melakukan plotting data sinyal
plt.plot(t,d)                                  # Data yang akan di plot yaitu data "d" terhadap perubahan "t"
plt.xlabel('t')                                # Label dari sumbu x
plt.ylabel('d(t)')                             # Label dari sumbu y

# Memberikan judul grafik
plt.suptitle("Grafik 4. Dirac Delta Function", fontsize = 20, y=1)

# Menampilkan grafik
plt.show()
```

```python
# Mengimport module yang digunakan dan mendefinisikan dengan alias yang lebih singkat
import numpy as np                             # Digunakan untuk membuat objek kelas array tunggal dan multidimensi.
import pylab as plt                            # Digunakan untuk visualisasi data

# Interval waktu dt untuk resolusi grafik
dt=0.1

# Interval waktu pertama 
t1=np.arange(-5,-2.1+dt,dt)                    # Saat t = -5 hingga t= -2.1

# Waktu kedua
t2=[-2.]                                       # Saat t = -2 

# Interval waktu ketiga
t3=np.arange(-1.9,0.9+dt,dt)                   # Saat t = -1.9 hingga t= 0.9

# Waktu keempat
t4=[1.]                                        # Saat t = 1

# Interval waktu kelima
t5=np.arange(1.1,10+dt,dt)                     # Saat t = 1.1 hingga t= 10

# Definisi nilai fungsi d(t) saat t1
d1=np.zeros_like(t1)                           # d(t) bernilai 0

# Definisi nilai fungsi d(t) saat t2
d2=[1.]                                        # d(t) bernilai 1

# Definisi nilai fungsi d(t) saat t3
d3=np.zeros_like(t3)                           # d(t) bernilai 0

# Definisi nilai fungsi d(t) saat t4
d4=[1.]                                        # d(t) bernilai 1

# Definisi nilai fungsi d(t) saat t5
d5=np.zeros_like(t5)                           # d(t) bernilai 0

# Gabungan dari variabel t dan d
t = np.concatenate((t1,t2,t3,t4,t5))
d = np.concatenate((d1,d2,d3,d4,d5))

# Melakukan plotting data sinyal
plt.plot(t,d)                                  # Data yang akan di plot yaitu data "d" terhadap perubahan "t"
plt.xlabel('t')                                # Label dari sumbu x
plt.ylabel('d(t)')                             # Label dari sumbu y

# Memberikan judul grafik
plt.suptitle("Grafik 5. Dirac Delta Function", fontsize = 20, y=1)

# Menampilkan grafik
plt.show()


```
